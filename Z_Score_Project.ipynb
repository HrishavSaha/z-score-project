```python
import statistics
import plotly.figure_factory as ff
import plotly.graph_objects as go
import pandas as pd
import random
import termcolor
```

File Reading:

```python
population_data = pd.read_csv("medium_data1.csv")['reading_time']
sample_data = pd.read_csv("read_time_sample.csv")['reading_time']

population_mean = statistics.mean(population_data)
sample_mean = statistics.mean(sample_data)
```

Functions:

```python
def sampler(data, iterations):
  temp = data.tolist()
  sample = []

  for i in range(0, iterations):
    rand = random.randint(0, len(temp)-1)
    sample.append(temp[rand])
    temp.pop(rand)

  mean = statistics.mean(sample)
  return mean

def setup(data, iterations, child_iterations):
  mean_list = []

  for i in range(0, iterations):
    mean = sampler(data, child_iterations)
    mean_list.append(mean)

  return mean_list

def plot_graph(data, sample, name, show_hist, ceil):
  mean = statistics.mean(data)
  stdev = statistics.stdev(data)
  sample_data_mean = statistics.mean(sample)

  if sample_data_mean > mean:
    print(termcolor.colored(("Intervention Successful!"), "green"))
  else:
    print(termcolor.colored(("Intervention Failed!"), "red"))

  plotter = ff.create_distplot([data], [name], show_hist=show_hist)

  plotter.add_trace(go.Scatter(x=[mean, mean], y=[0, ceil], mode='lines', name='Mean'))
  plotter.add_trace(go.Scatter(x=[mean-(3*stdev), mean+(3*stdev)], y=[ceil, ceil], mode='lines', name='StDev 3'))
  plotter.add_trace(go.Scatter(x=[mean-(2*stdev), mean+(2*stdev)], y=[ceil-0.02, ceil-0.02], mode='lines', name='StDev 2'))
  plotter.add_trace(go.Scatter(x=[mean - stdev, mean + stdev], y=[ceil-0.04, ceil-0.04], mode='lines', name='StDev 1'))
  
  plotter.add_trace(go.Scatter(x=[sample_data_mean, sample_data_mean], y=[0, ceil], mode='lines', name='Sample Mean'))

  plotter.show()

def z_score(population, sample):
  population_mean = statistics.mean(population)
  sample_mean = statistics.mean(sample)
  population_stdev = statistics.stdev(population)

  z_score = (sample_mean - population_mean)/population_stdev

  return z_score
```

Plots:

```python
mean_list = setup(population_data, 100, 30)
plot_graph(mean_list, sample_data, "Reading Time", False, 0.8)
```

```python
plot_graph(population_data, sample_data, "Reading Time", False, 0.2)
```

```python
z = z_score(population_data, sample_data)
if z > 0:
  print(termcolor.colored(("Z Score:", z), 'green'))
else:
  print(termcolor.colored(("Z Score:", z), 'red'))
```
